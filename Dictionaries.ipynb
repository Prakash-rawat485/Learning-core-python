```python
d={'name':'Prakash','age':23,'Salary':500000,'Place of birth':'Haldwani'}            # direct call
print(d['name'])
```

# .get()

```python
print(d.get('age'))               #  used .get()
```

```python
print(d.get('Salary'))
```

```python
print(d.get('Occupation'))
```

```python
d['Occupation']='Chief data scientist'
print(d)
```

# Update() :-

```python
d.update({'name':'Deepak rawat','Pin code':263139})
print(d)
```

```python
d.update({'country':'India'})
print(d)
```

# pop() :-


```python
popped=d.pop('country')
print(popped)
print(d)
```

# popitem() :-

```python
popped2=d.popitem()                                            # delete the last element of the dict
print(popped2)
print(type(popped2))
print(d)
```

```python
d.pop("Haldwani")                                       # takes only keys not values.
print(d) 
```

# keys(),values(),Items() :-

```python
c={'name':'Prakash','age':23,'Salary':500000,'Place of birth':'Haldwani'} 
c.keys()
```

```python
print(type(c))
```

**As class dict is not useful to us so we change the data type through type casting:-**

```python
c_list=list(c.keys())
print(c_list)
print(type(c_list))
```

```python
Values=c.values()
print(Values)
print(type(Values))
```

```python
values_tuple=tuple(c.values())
print(values_tuple)
print(type(values_tuple))
```

```python
Items=c.items()
print(Items)
print(type(Items))
```

```python
Item_set=set(c.items())
print(Item_set)
print(type(Item_set))
```

```python
set4=set(c)
print(set4)
```

```python
list1=list(c)
print(list1)
```
