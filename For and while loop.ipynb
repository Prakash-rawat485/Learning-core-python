# For loop:
- **A for loop is used to reappear a block of code a specific number of times,or for each item in a sequence(like a list,stirng,etc).**

#Syntax

**for variable in sequence:**

```python
fruits=['apple','banana','cherry']
for fruits in fruits:
    print(fruits)
```

```python
fruits=['apple','banana','cherry']
for fruits in fruits:
    print(fruits,end=' ')
```

```python
fruits=['apple','banana','cherry']
for fruits in fruits:
    print(fruits,end='*****')
```

```python
fruits=['apple','banana','cherry']
for fruits in fruits:
    print(fruits,end=('tasty*').split('*'))
```

```python
fruits=['apple','banana','cherry']
for fruits in fruits:
    print(fruits,end=(' tasty*\n'))
```

```python
fruits=['apple','banana','cherry']
for fruits in fruits:
    print(fruits,end=(' tasty*\n\t'))
```

```python
fruits=['apple','banana','cherry']
for fruits in fruits:
    print(fruits)
    end=(' tasty*\n\t')
```

```python
fruits=['apple','banana','cherry']
for fruits in fruits:
    print(fruits,end=('fruits.append('blueberry'))
```

```python
fruits=['apple','banana','cherry']
for fruits in fruits:
    print(fruits,'fruits.append('guava'))
```

# Using range() with for loop:
-range(start,stop,step) is often used with for to repeat something multiple times:

```python
for i in range(12,41,2):
    print(i)
```

# Reverse range:

```python
for i in range(6,0):
    print(i)
```

```python
for i in range(6,0,-1):
    print(i)
```

# nested for loop:
- A nested loop is when one for loop is inside another.

```python
for i in range(3):
    for j in range(2):
        print(f'i={i},j={j}')
```

# WHILE LOOP:
- A while loop repeats a block of code as long as a condition in true.

```python
i=1
while i<=5:
    print(i)
    i=i+1
print('get up it is time to study')
```

```python
i=1
while i<=5:
    print(i)
    i=i+1
```

```python
i=1
while (i<=20):
    print(i)
    i=i+1
print('yeah boy ')
```

```python
i=1
while (i<=20):
    print(i)
    i=i*2
print('yeah boy ')
```

```python
i=1
while i>=-20:
    print(i)
    i=i-1
print('the loop ended')
```

```python
i=-2
while i>=-20:
    print(i)
    i=i*2
print('hey there')
```

```python
r=10
while r>=5:
    print(r)
    r=r-1
print('the loop has to end now')
```

```python
w=2
while w<=25:
    print(w)
    w=w*2
print('end')
```

```python
t=400
while t>=50:
    print(t)
    t=t/2
print('hmm ended')
```

# crux of making patterns.
- **outer loop controls the number of rows(vertical dimension).**

- **inner loop controls character per row(horizontal dimension).**

- **print(...,end='') keeps printing in the same line.**

- **print()moves to the next line after a row is done.**

- **logic in inner decides what to print:*,number,space,etc>**

- **conditions(if)used for shapes like hollow boxes,diagonals etc.**


```python
row=10
for i in range(1,row+1):
    for j in range(2,i+1):
        print(j,end=' ')
    print()
```

```python
row=10
for i in range(1,row+1):
    print('*'*i)
```

1. **print numbers in triange form:**

```python
rows=5
i=1
while i<=rows:
    j=1
    while j<=i:
        print(j,end=' ')
        j+=1
    print()
    i+=1
```
