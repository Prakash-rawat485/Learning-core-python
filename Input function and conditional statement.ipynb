# get multiple inputs from a user in one line: use of split() function:-
- It is possible in python to get multiple values from the user.
- take each input separated by space split input using split() get the value of individual  input.
- **common use of split():**
- tokenizing a string: if you have a string that contains multiple words or values separated by a specific character(such as comma,space or any other delimiter) you can use split() to split the string into individual tokend.
- **split() takes only string.**
- **, comes as default.**
- **helps to remove any delimiter.**

```python
sentence='helloo how are you today'
words=sentence.split()
print(words)
```

```python
type(words)
```

```python
sentence='helloo how are you today'
words=sentence.split('*')
print(words)
```

```python
sentence='helloo how are you today'
words=sentence.split(1)
print(words)
```

#  Extracting substring:
- if you have a string with a consistent pattern and you want to extract specific parts of it, you can use split() to separate the string into parts and access the desired substring.

```python
date='2023*02*02'
year,month,day=date.split('*')
print(year,month,day)
```

```python
date='2023*02*02'
newdate=list(date.split('23'))
print(newdate)
```

```python
no1=float(input('Enter first number:'))
no2=float(input('Enter second number:'))
operator=input('operator to be used:')
if operator=='+':
    output=no1+no2
    print(output)
elif operator=='-':
    output=no1-no2
    print(output)
elif operator=='*':
    output=no1*no2
    print(output)
elif operator=='/':
    output=no1/no2
    print(output)
else:
    print('Invalid operator')                                # the print(output) is to be given before the else statement,otherwise the error will be shown.
 
```
