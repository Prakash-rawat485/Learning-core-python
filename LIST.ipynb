# difference between append and extend
- **append return the same string whereas extend return the subtring of the given string.**

- **append can take int,whereas extend does not take int.**

- **in append if the list is passed than in the output the list is separetly shown inside the previous listi.e nested list,but in extend the whole result is showed as one**

- **.extend works for iterable data types-set,tuple etc**

# Removing elements from  the list.
- **remove(item): to remove the first occurence of the item from the list.**
- **pop(index): removes and returns the item at the given index from the list.**
- **clear(): to remove all the items from the list, the output will be an empty list.**
- **del list_name: delete the entire list.**

# remove: takes only one argument,and removes the first occuring item.

```python
hm=[12,77,45,65,77,89,'prakash rawat']
```

# Concatenation of  two list 

```python
my_list1=[1,2,3]
my_list2=[4,5,6]
my_list1.extend(my_list2)
print(my_list1)

# using + operator

my_list3= my_list1 + my_list2
print(my_list3)
```

```python
new_1=[23,42,41,51,56]
new_2=[67,87,34,63,23]
new_3=new_1+new_2
print(new_3)
```
