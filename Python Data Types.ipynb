-list

# List(mutuble/unhashable/sequence data type):-
- it is a mutable data type having following characterstics.
- It contains **duplicate items**.
- It can be modified,replaced or deleted **without changing memory addressed**.
- List is **ordered** i.e the order of the element is based in such a way that they can be edit in some order.
- The items of the list can be accessed directly ********using their index number and the index number starts from zero(0)********.
- The list may contains *mixed type of data*.
- **[] represents the list**.
- The list can contain **duplicates** i.e list can have two items with the same values.
- **List is iterable**

# Iterating a list.
- The objects in the list can be iterated over one by one by using a for loop.

```python
my_list=[4,5,'Geoffrey Hinton',2.5,'Sam altman']
print(my_list)       # without iteration
```

***iterate a list***

```python
for i in my_list:
    print(i)
```
