# Identity operators in python.
- In python **is and not is**  are the identity operators both are used to check if two values are located in same memory part,if two variable are equal it does not means that they are identical.

```python
a=15
print(id(a))
```

```python
b=15
print(id(b))
```

```python
a is b                                         # it tell about the identity
```

```python
x=15
print(id(x))
```

```python
y=90
print(id(y))
```

```python
x is y                                      # here both have different memory
```

```python
x is not y
```

```python
t=15.999
print(id(t))
```

```python
o=42.999
print(id(o))
```

```python
t is o
```

```python
t is not o
```

```python
n=1.2
w=1.22
print(id(n)),print(id(w))
```
