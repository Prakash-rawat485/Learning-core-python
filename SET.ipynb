# SET:
- **Unordered:** the items in the set are unordered unlike list it will not maintain the order in which the items are inserted.there will be no index value assigned to each item in the set
- **Unchangeable:** set items must be immutable.We cannot change the set items. But set as as a whole is mutable.
- **unique:** There cannot be two items with the same value in the set,i.e **no duplicates** 
- **inside the set the mutables cannot come like list,set and dict, only the immutable can come  there like int,string,tuple.**

```python
list1=[2,5,2,5,7,8]
new_set=set(list1)
print(new_set)
```

```python
new_list=list(new_set)                                     # type casting
print(new_list)
```

# Removing items from a set.
- **Remove()** - removes a single item . it takes one parameter i.e items to be removed.
- **discard()**- To remove a single item that may or may not be present in the set. Takes one parameter.if the item is present it will remove it. and if not if will not throw an error.
- **pop()**- to remove any reandom item from a set
- **clear()**- to remove all items from the set. the output will be an empty set.
- **del set**- delete the entire set.

```python
set2={'ab','prakash','sakshi',23,42}
print(id(set2))
set2.remove('ab')
print(set2)
print(id(set2))
```

```python
set2.remove(88)
print(set2)
```

```python
set2.discard(42)
print(set2)
```

```python
set2.discard('vishal')                      # item is not present but it is not throwing any error it is giving the same output.
print(set2)
```

```python
set2.discard(23)
print(set2)
```

# Set operations:- 
- **Union**- All the items of both sets will be returned,only the duplicate items will be dropped.
- **Intersection**- It will return the element which is common in both the set.
- **Difference**- this return the unique elements of the first set which are not in second set.
- **symmetric_difference**- return the elements of  both sets which are not common.

```python
set5={42,52,62,63,98,67,'23'}
set6={62,45,23,98}
set5.union(set6)
```

```python
set5.intersection(set6)
```

```python
n=set5&set6
print(n)
```

```python
set5.difference(set6)
```

```python
set5.symmetric_difference(set6)
```
