# Membership

```python
# Membership operators: in    and    not in
t=(2,5,6)
print(3 in t)
print(6 not in t)
```

```python
print(2 in t)
print(t in 2) # illogical
```

```python
name1=('prakash','rawat')
name2=(('prakash','rawat'),'Sam','Bill')
print(name1 in name2)
print(name2 in name1)
print(name1[1] in name2)
print('Sam' in name2)
print(('prakash','rawat') in name2)
```
